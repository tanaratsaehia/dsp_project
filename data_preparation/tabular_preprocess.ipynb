```python
import os
import pandas as pd
import re
from glob import glob
from preprocess_tabular import PreprocessAccel

pre_acc = PreprocessAccel(sampling_rate=50)

root_path = r'D:\git_repository\dsp_project\ignore_dir\datasets\clean_datasets'
subfolder = ["hand", "pocket"]
magnitude_column_name = "magnitude"  # Adjust if needed
window_size = 5  # Window size in seconds
overlap = 0.5

# Define the folder path


# Function to extract label from filename
def extract_label(filename):
    match = re.match(r'([a-zA-Z_]+)', filename)
    return match.group(1) if match else filename

# Process all CSV files in the folder
all_data = []
for sub in subfolder:
    folder_path = os.path.join(root_path, sub)
    for file_path in glob(os.path.join(folder_path, "*.csv")):
        filename = os.path.basename(file_path)
        label = extract_label(filename)

        data = pd.read_csv(file_path)
        processed_data = pre_acc.segment_and_flatten_magnitude(
            label=label, magnitude_column_name=magnitude_column_name, 
            df=data, window_size_sec=window_size, overlap=overlap
        )
        all_data.append(processed_data)

# Combine all processed data
merged_df = pd.concat(all_data, ignore_index=True)
```

```python
root_path = r'D:\git_repository\dsp_project\ignore_dir\datasets\ready_to_train'
merged_df.to_csv(os.path.join(root_path, f"cleaned_combine_all_class_with_label_no_fft_win{str(window_size)}_lab{str(int(overlap*100))}.csv"))
```
